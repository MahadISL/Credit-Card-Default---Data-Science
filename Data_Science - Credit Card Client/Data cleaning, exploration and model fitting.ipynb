```python
X_train, X_test, y_train, y_test = train_test_split(
df['LIMIT_BAL'].values.reshape(-1,1), df['default payment next month'].values, test_size=0.2, random_state=20)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
my_lr = LogisticRegression()
```

```python
my_lr.fit(X_train, y_train)
```

```python
y_prob_pred = my_lr.predict_proba(X_test)
y_prob_pred
```

```python
prob_sum = np.sum(y_prob_pred, 1)
prob_sum
```

```python
pos_prob = y_prob_pred[:,1]
pos_prob
```

```python
fpr, tpr, thresholds = metrics.roc_curve(y_test, pos_prob)
```

```python
plt.plot(fpr, tpr, '*-')
plt.plot([0,1],[0,1], 'r--')
plt.legend(['Logistic Regression', 'Random Chance'])
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC curve')
```

```python
metrics.roc_auc_score(y_test, pos_prob)
```

```python
precision, recall, thres = \
metrics.precision_recall_curve(y_test, pos_prob)
```
